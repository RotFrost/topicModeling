```python
import pandas as pd
import bokeh
import locale
import os
import json
import numpy as np
locale.setlocale(locale.LC_ALL, 'de_DE')
```

```python
def getResultDict():
    filePath = os.path.join('result', 'resultRecordDict.json')
    if os.path.exists(filePath):
        with open(filePath) as json_file:
            dict_ = json.load(json_file)
        return dict_
    else:
        return dict()
```

```python
dict_ = getResultDict()
```

```python
df = pd.DataFrame.from_dict(dict_, orient='index')
```

```python
df.columns
```

## Beste Models nach den coherence score u_mass

```python
df[['modelType', 'gensimPreProcess', 'allowedTags', 'perplexity','coherence','topicN','coherenceCV', 'topicList']].sort_values(by='coherence', ascending=False)[0:20]
```
