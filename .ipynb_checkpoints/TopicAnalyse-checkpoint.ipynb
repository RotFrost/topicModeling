# Visualisierung

```python
from bokeh.plotting import figure, output_file, show
```

```python
dfy = df.groupby(df['time'].dt.year).count()
dfy_ = dfy.reset_index(drop=True)
time = pd.to_datetime(dfy.index, format="%Y") 
dfy_zur = df.loc[df['richText'].str.contains('Zur externen Meldung')].groupby(df['time'].dt.year).count().reset_index(drop=True)
dfy_Minuszur = df.loc[~df['richText'].str.contains('Zur externen Meldung')].groupby(df['time'].dt.year).count().reset_index(drop=True)
```

```python
p = figure(title="Anzahl der Dokumente über die Jahre", x_axis_type="datetime", plot_width=800, plot_height=500)
p.line(x=time, y=dfy_['title'], legend_label='Gesamt')
p.line(x=time, y=dfy_zur['title'], legend_label='Ohne Haupttext',  line_color="red")
p.line(x=time, y=dfy_Minuszur['title'], legend_label='Mit Haupttext', line_color="green")
p.yaxis.axis_label = "Anzahl der Dokumente"
p.xaxis.axis_label = "Jahre"
show(p)
```
